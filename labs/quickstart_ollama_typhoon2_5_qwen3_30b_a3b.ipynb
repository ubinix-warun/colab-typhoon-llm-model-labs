```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

model_id = "scb10x/typhoon2.5-qwen3-30b-a3b"

# --- กำหนดค่าการบีบอัดโมเดล (4-bit Quantization) ---
# นี่คือส่วนที่ช่วยลดการใช้ VRAM และป้องกัน CUDA error
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# --- โหลด Tokenizer (ล่ามแปลภาษา) ---
tokenizer = AutoTokenizer.from_pretrained(model_id)

# --- โหลด Model (สมองกล AI) ---
# เราจะส่งค่า quantization_config เข้าไปตรงนี้
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=quantization_config,
    torch_dtype=torch.bfloat16, # ช่วยให้โมเดลทำงานกับ GPU T4 ได้ดีขึ้น
    device_map="auto" # ช่วยจัดการส่งส่วนต่างๆ ของโมเดลไปที่ GPU อัตโนมัติ
)
```

```python
from transformers import pipeline

# สร้าง pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=512, # เพิ่มความยาวสูงสุดของคำตอบเผื่อไว้
    do_sample=True,
    temperature=0.7,
    top_p=0.95
)

# เตรียมคำสั่ง (Prompt) ในรูปแบบเฉพาะของโมเดล
prompt = "<s>[INST] สวัสดี ประเทศไทยมีทั้งหมดกี่จังหวัด? และช่วยแนะนำจังหวัดที่น่าเที่ยวมา 3 แห่งพร้อมเหตุผลสั้นๆ [/INST]"

# สั่งให้ pipeline ทำงาน
result = pipe(prompt)

# พิมพ์ผลลัพธ์
print(result[0]['generated_text'])
```
