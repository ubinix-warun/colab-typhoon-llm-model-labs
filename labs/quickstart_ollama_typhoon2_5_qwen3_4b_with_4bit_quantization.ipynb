```python
import torch
import time
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline

# --- ติดตั้ง Library (หากยังไม่ได้ติดตั้ง) ---
# !pip install -q -U transformers accelerate bitsandbytes

model_id = "scb10x/typhoon2.5-qwen3-4b"

# --- 1. กำหนด Config ทั้ง 3 แบบ ---
config_8bit = BitsAndBytesConfig(
    load_in_8bit=True
)

config_4bit_nf4 = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

config_4bit_fp4 = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="fp4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# --- 2. ฟังก์ชันสำหรับ Benchmark ---
def run_benchmark(config_name, quantization_config):
    print(f"\n--- 🚀 กำลังทดสอบ: {config_name} ---")

    # 1. โหลดโมเดล (ใช้เวลาตรงนี้)
    start_load_time = time.perf_counter()
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        quantization_config=quantization_config,
        torch_dtype=torch.bfloat16, # bfloat16 จำเป็นสำหรับ 4-bit compute
        device_map="auto"
    )
    end_load_time = time.perf_counter()
    print(f"เวลาโหลดโมเดล: {end_load_time - start_load_time:.2f} วินาที")

    # 2. สร้าง Pipeline
    pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

    # 3. เตรียม Prompt (ใช้ Chat Template ที่ถูกต้อง)
    messages = [{"role": "user", "content": "ช่วยเขียนบทสรุปสั้นๆ เกี่ยวกับประโยชน์ของ AI ใน 5 ข้อ"}]
    prompt_text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    tokens_to_generate = 128

    # 4. WARM-UP (สำคัญมาก!)
    print("กำลัง Warm-up...")
    _ = pipe(prompt_text, max_new_tokens=10, pad_token_id=tokenizer.eos_token_id)

    # 5. เริ่มวัดผลจริง
    torch.cuda.empty_cache()
    torch.cuda.reset_peak_memory_stats()

    start_inference_time = time.perf_counter()

    result = pipe(
        prompt_text,
        max_new_tokens=tokens_to_generate,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=False # ปิดการสุ่มเพื่อผลลัพธ์ที่คงที่
    )

    torch.cuda.synchronize() # รอให้ GPU ทำงานเสร็จ
    end_inference_time = time.perf_counter()

    # 6. รวบรวมผลลัพธ์
    time_taken = end_inference_time - start_inference_time
    tokens_per_sec = tokens_to_generate / time_taken
    peak_vram_bytes = torch.cuda.max_memory_allocated()
    peak_vram_gb = peak_vram_bytes / (1024**3)

    print(f"--- ผลลัพธ์ {config_name} ---")
    print(f"VRAM สูงสุดที่ใช้: {peak_vram_gb:.2f} GB")
    print(f"เวลาสร้าง {tokens_to_generate} โทเค็น: {time_taken:.2f} วินาที")
    print(f"ความเร็ว: {tokens_per_sec:.2f} Tokens/Second")
    print("\n--- คุณภาพคำตอบ (ตัวอย่าง) ---")
    print(result[0]['generated_text'])

    # คืนค่า VRAM เพื่อรัน Config ถัดไป
    del model
    del pipe
    torch.cuda.empty_cache()

# --- 3. รันการทดสอบ! ---
# (หมายเหตุ: ควร Restart Runtime ก่อนรันแต่ละตัว เพื่อป้องกัน VRAM ค้าง)
# run_benchmark("8-bit", config_8bit)
# run_benchmark("4-bit NF4", config_4bit_nf4)
run_benchmark("4-bit FP4", config_4bit_fp4)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

model_id = "scb10x/typhoon2.5-qwen3-4b"

# --- กำหนดค่าการบีบอัดโมเดล (4-bit Quantization) ---
# นี่คือส่วนที่ช่วยลดการใช้ VRAM และป้องกัน CUDA error
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# --- โหลด Tokenizer (ล่ามแปลภาษา) ---
tokenizer = AutoTokenizer.from_pretrained(model_id)

# --- โหลด Model (สมองกล AI) ---
# เราจะส่งค่า quantization_config เข้าไปตรงนี้
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=quantization_config,
    torch_dtype=torch.bfloat16, # ช่วยให้โมเดลทำงานกับ GPU T4 ได้ดีขึ้น
    device_map="auto" # ช่วยจัดการส่งส่วนต่างๆ ของโมเดลไปที่ GPU อัตโนมัติ
)
```

```python
from transformers import pipeline

# สร้าง pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=512, # เพิ่มความยาวสูงสุดของคำตอบเผื่อไว้
    do_sample=True,
    temperature=0.7,
    top_p=0.95
)

# เตรียมคำสั่ง (Prompt) ในรูปแบบเฉพาะของโมเดล
prompt = "<|im_start|> สวัสดี ประเทศไทยมีทั้งหมดกี่จังหวัด? และช่วยแนะนำจังหวัดที่น่าเที่ยวมา 3 แห่งพร้อมเหตุผลสั้นๆ <|im_end|>"

# สั่งให้ pipeline ทำงาน
result = pipe(prompt)

# พิมพ์ผลลัพธ์
print(result[0]['generated_text'])
```
